# Starbucks Dataset Analysis

```python
%pip install pandas seaborn matplotlib -q
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/starbucks.csv')
df.columns = df.columns.str.strip()
df['Total Fat (g)'] = df['Total Fat (g)'].str.replace(' ', '.').astype(float)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby('Beverage_category')['Calories'].mean().plot(kind='barh', color='teal')
```

Smoothies have the most calories while regular black coffee has the least.

```python

sns.scatterplot(x='Total Fat (g)', y='Saturated Fat (g)', data=df)
```

The more total fat a drink has, the more saturated fat it has in most cases.

```python
sorted_df = df.sort_values(by='Caffeine (mg)', ascending=True)
sorted_df['Caffeine (mg)'] = pd.to_numeric(sorted_df['Caffeine (mg)'], errors='coerce')
sns.histplot(y='Caffeine (mg)', data=sorted_df)
# plt.xlabel('Caffeine (mg)')
# plt.ylabel('Frequency')
plt.title('Distribution of Caffeine Content')
```

Many beverages have 0-200 mg of caffeine, and only a few have 200-400 mg.

```python
sns.set()
fig, axs = plt.subplots(1, 3)

sns.boxplot(ax=axs[0], x='Beverage_category', y='Protein (g)', data=df)
sns.boxplot(ax=axs[1], x='Beverage_category', y='Dietary Fibre (g)', data=df)
sns.boxplot(ax=axs[2], x='Beverage_category', y='Sugars (g)', data=df)
axs[0].tick_params(axis='x', rotation=90)
axs[1].tick_params(axis='x', rotation=90)
axs[2].tick_params(axis='x', rotation=90)
# axs[0].set_title('Protein (g)')
# axs[1].set_title('Dietary Fibre (g)')
# axs[2].set_title('Sugars (g)')



# for i, nutrient in enumerate('Protein (g)', 'Dietary Fibre (g)', 'Sugar (g)'):
#     pass

plt.show()
```

```python

df['Vitamin A (% DV)'] = df['Vitamin A (% DV)'].astype(str).str.replace('%', '').astype(float)
df['Vitamin C (% DV)'] = df['Vitamin C (% DV)'].astype(str).str.replace('%', '').astype(float)
df['Calcium (% DV)'] = df['Calcium (% DV)'].astype(str).str.replace('%', '').astype(float)
df['Iron (% DV)'] = df['Iron (% DV)'].astype(str).str.replace('%', '').astype(float)
info = df.groupby('Beverage_prep')[['Vitamin A (% DV)', 'Vitamin C (% DV)', 'Calcium (% DV)', 'Iron (% DV)']].mean()
ax = info.plot.barh(stacked=True)
plt.title('Vitamin and Mineral Contribution by Beverage Preparation')
plt.xlabel('Beverage Category')
plt.ylabel('% DV')
plt.xticks(rotation=90)
plt.show()
```

Most beverages provide a large percentage of a person's daily value of Calcium, but much less Vitamin C and Iron.
